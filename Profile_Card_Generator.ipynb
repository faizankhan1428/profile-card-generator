```python
from google.colab import files

uploaded = files.upload()
photo_path = list(uploaded.keys())[0]  # get uploaded file name
```

```python
# Input details
name = input("Full Name: ")
user_id = input("ID: ")
nationality = input("Nationality: ")
job_title = input("Job Title: ")
```

```python
from PIL import Image, ImageDraw, ImageFont
import uuid
import os # Import the os module to check for file existence

# Create card base
card_width, card_height = 1000, 600
card = Image.new("RGB", (card_width, card_height), "#1e1e2f")
draw = ImageDraw.Draw(card)

# Add top & bottom stripe
draw.rectangle([0, 0, card_width, 100], fill="#00c896")     # top stripe
draw.rectangle([0, card_height-50, card_width, card_height], fill="#00c896")  # bottom

# Add title
# Check if the specific font exists, otherwise use a common alternative
title_font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf"
if os.path.exists(title_font_path):
    title_font = ImageFont.truetype(title_font_path, 45)
else:
    # Use a commonly available font like Arial if DejaVuSans-Bold is not found
    print(f"Warning: Font '{title_font_path}' not found. Using Arial.ttf instead.")
    try:
        # Common paths for Arial in Colab
        arial_font_path = "/usr/share/fonts/truetype/liberation/LiberationSans-Bold.ttf" # Liberation Sans is often used as an Arial equivalent
        if not os.path.exists(arial_font_path):
             arial_font_path = "/usr/share/fonts/truetype/msttcorefonts/Arial_Bold.ttf" # Another possible path for Arial Bold
        title_font = ImageFont.truetype(arial_font_path, 45)
    except Exception as e:
        print(f"Error loading alternative font: {e}")
        # Fallback to a basic Pillow font if true type fails
        title_font = ImageFont.load_default()


draw.text((40, 25), "PROFILE CARD", font=title_font, fill="white")

# Load user photo
photo = Image.open(photo_path).convert("RGB").resize((180, 180))
card.paste(photo, (800, 40))

# Text info
# Check if the specific font exists for info, otherwise use a common alternative
info_font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf"
if os.path.exists(info_font_path):
    info_font = ImageFont.truetype(info_font_path, 32)
else:
    # Use a commonly available font like Arial if DejaVuSans is not found
    print(f"Warning: Font '{info_font_path}' not found. Using Arial.ttf instead.")
    try:
        # Common paths for Arial in Colab
        arial_font_path = "/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf" # Liberation Sans is often used as an Arial equivalent
        if not os.path.exists(arial_font_path):
             arial_font_path = "/usr/share/fonts/truetype/msttcorefonts/Arial.ttf" # Another possible path for Arial
        info_font = ImageFont.truetype(arial_font_path, 32)
    except Exception as e:
         print(f"Error loading alternative font: {e}")
         # Fallback to a basic Pillow font if true type fails
         info_font = ImageFont.load_default()


y_start = 150
spacing = 60

draw.text((60, y_start),     f"Name: {name}",        font=info_font, fill="white")
draw.text((60, y_start+spacing), f"ID: {user_id}",       font=info_font, fill="white")
draw.text((60, y_start+2*spacing), f"Nationality: {nationality}", font=info_font, fill="white")
draw.text((60, y_start+3*spacing), f"Job Title: {job_title}", font=info_font, fill="white")
```

```python
from IPython.display import display
filename = f"profile_card_{uuid.uuid4().hex[:6]}.png"
card.save(filename)

display(card)
files.download(filename)
```
